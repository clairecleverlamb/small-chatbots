```python
import gradio as gr
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")
def analyze_sentiment(text):

    result = sentiment_pipeline(text)[0]
    label = result["label"]
    score = result["score"]
    return f"Sentiment: {label} (confidence: {score:.2f})"


demo = gr.Interface(
    fn=analyze_sentiment,
    inputs="text",
    outputs="text",
    title="Simple Sentiment Analysis",
    description="Type a sentence and see whether it's detected as POSITIVE or NEGATIVE."
)


demo.launch()
```

```python
from transformers import pipeline


summarizer = pipeline("summarization", model="facebook/bart-large-cnn")


text = """
Transformers are taking the machine learning world by storm.
They are particularly effective for tasks involving natural language processing,
such as text generation, summarization, and translation.
Many modern large language models, including GPT, are built on transformer architectures.
"""


summary = summarizer(
    text,
    max_length=50,
    min_length=25,
    do_sample=False
)


print("Original Text:\n", text)
print("\nSummary:\n", summary[0]['summary_text'])
```
